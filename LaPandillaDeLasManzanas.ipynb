Imagina que un grupo de amigos quiere saber más sobre su cosecha de manzanas.
Cosechan 10 manzanas y quieren analizar su peso para ver cómo es su producción este año.
Queremos calcular:
Media (promedio de peso)
Mediana (el elemento que está en el medio, el peso central)
Moda (el pero que más se repite)
Desviación estándar: qué tanto varían los pesos. Raiz cuadrada de la varianza.


```python
import numpy as np
import statistics as stats

manzanas = [150, 180, 200, 170, 150, 190, 200, 210, 150, 180]

#Estadísticas descriptivas
media = np.mean(manzanas) #mean = media o promedio
mediana = np.median(manzanas) #median = mediana
moda = stats.mode(manzanas) #moda = moda
desviación = np.std(manzanas) #variabilidad de los datos

#Impresión de resultados
print(f"Análisis de la cosecha de manzanas 🍎:"
print(f"Media: {media} gramos"))
print(f"Mediana: {mediana} gramos")
print(f"Moda: {moda} gramos")
print(f"desviación: {desviación} gramos")
```
